```python
#Summary statistics
# we split the data into two parts and calculate mean and variance

X = df.values
split = round(len(X)/2)
X1, X2 = X[0:split], X[split:]
mean1, mean2 = X1.mean(), X2.mean()
var1, var2 = X1.var(), X2.var()
print("Part 1 mean and variance: %d,  %d"%(mean1, var1))
print("Part 2 mean and variance: %d, %d"%(mean2, var2))
```
