```python
import pandas as pd
import numpy as np
```

```python
sfp_data = pd.read_csv('bp/Building_Permits.csv')
np.random.seed(0)
```

```python
sfp_data.sample(5)
```

```python
len(sfp_data)
```

```python
missing_values_count = sfp_data.isnull().sum()
missing_values_count
```

```python
#finding percentage of missing values

total_cells = np.product(sfp_data.shape)
total_missing_cells = missing_values_count.sum()

missing_percentage = (total_missing_cells/total_cells)*100
print("missing_cells percentage: ", missing_percentage)
```

```python
rows_with_missing_dropped = sfp_data.dropna(axis=1)
columns_with_missing_dropped = sfp_data.dropna()

print("Original shape of data:", sfp_data.shape)
print("Shape after dropping rows with missing values:", rows_with_missing_dropped.shape)
print("Shape after dropping columns with missing values:", columns_with_missing_dropped.shape)

print(rows_with_missing_dropped.head())
print(columns_with_missing_dropped.head())
```

```python
#replacing all missing values with 0
subset_sfp_data = sfp_data.loc[:, 'Permit Type' : 'Issued Date'].head()

subset_sfp_data
```

```python
subset_sfp_data.fillna(value=0)
subset_sfp_data
```

```python
print("Missing values:", missing_values_count.sum())
ins_zero_data = sfp_data.fillna(0)
new_missing_number = ins_zero_data.isnull().sum()
print("Missing values after insertion of zero:", new_missing_number.sum())
```

```python
ins_bfill_data = sfp_data.fillna(method='bfill', axis=0).fillna(0)
print("Missing values after insertion by bfill method:", ins_bfill_data.isnull().sum().sum())
ins_bfill_data.head()
```
